```python
ax = sns.countplot(x='Attrition', data=df)
total = len(df)
for bars in ax.containers:
    ax.bar_label(bars, labels=[f'{(v.get_height()/total)*100:.2f}%' for v in bars])
plt.title('Attrition Distribution with Percentage')
plt.xlabel('Attrition')
plt.ylabel('Percentage')
plt.show()
```

```python
avg_tenure=df['YearsAtCompany'].mean()
print(f'Average years of employee to leave the company is {avg_tenure} years')
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['ï»¿Age'], kde=True, color='blue')
plt.title('Distribution of Employees by Age')
plt.xlabel('Age')
plt.ylabel('Count')

```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', hue='Gender', data=df, palette='Set2', legend=False)
plt.title('Distribution of Employees by Gender')
plt.xlabel('Gender')
plt.ylabel('Employee Count')

```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', data=df, palette='Set2', hue='Department', legend=False)
plt.title('Distribution of Employees by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
df_attrition = df[df['Attrition'] == 'Yes' ]
df_attrition.head()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='JobRole', hue='Attrition', data=df)
plt.title('Attrition by Job Role')
plt.xticks(rotation=90)

```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Education', y='MonthlyIncome', hue='Attrition', data=df)
plt.title('Monthly Income by Education Level and Attrition')

```

```python
attrition_counts = df['Attrition'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue'])
plt.title('Overall Attrition Rate')

```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='JobRole', y='DistanceFromHome', hue='Attrition', data=df, palette='Set2')
plt.title('Distance from Home by Job Role and Attrition')
plt.xticks(rotation=90)

```

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
sns.kdeplot (data=df_attrition,x='ï»¿Age' ,fill=True ,ax=axes)
axes.set_title('Attrition by Age')
axes.set_xlabel('Age')
axes.set_ylabel('Density')           
             
```

```python
attrition_by_gender = df[df['Attrition'] == 'Yes'].groupby('Gender').size() / df.groupby('Gender').size() * 100
attrition_by_gender = attrition_by_gender.reset_index(name='Attrition Rate (%)')
plt.figure(figsize=(6, 4))
sns.barplot(x='Gender', y='Attrition Rate (%)', data=attrition_by_gender, palette='Set2', hue='Gender', legend=False)
plt.title('Attrition Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Attrition Rate (%)')


```

```python
worklife_attrition = pd.crosstab(df['WorkLifeBalance'], df['Attrition'])
worklife_attrition.plot(kind='bar', stacked=True, figsize=(8,4))
plt.title('WorkLifeBalance Breakdown by Attrition')
plt.xlabel('WorkLifeBalance')
plt.ylabel('Number of Employees')
plt.show()
```
